```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset/cleaned_data.csv')
df.head()
```

```python
# Get columns with int or float dtypes
num_cols = df.select_dtypes(include=['int', 'float']).columns
# Display numerical summaries
df[num_cols].describe()
```

```python
# Numerical summaries of age by diagnosis group
summary = df.groupby('diagnosis')['age'].describe()
print(summary)
```

```python
# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['age'], bins=15, color='#3498db', edgecolor='black', alpha=0.8)

# Add titles and labels
plt.title('Distribution of Age Among Patients', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Number of Patients', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```

```python
# Define color theme
colors = ['#3498db', '#e74c3c']  # Blue (No Alzheimerâ€™s), Red (Alzheimerâ€™s)


# Create figure
plt.figure(figsize=(10, 6))

# Plot histograms for each diagnosis group
plt.hist(
    df[df['diagnosis'] == 0]['age'],
    bins=15,
    alpha=0.6,
    label="No Alzheimer's (0)",
    color=colors[0],
    edgecolor='black'
)
plt.hist(
    df[df['diagnosis'] == 1]['age'],
    bins=15,
    alpha=0.6,
    label="Alzheimer's (1)",
    color=colors[1],
    edgecolor='black'
)

# Add labels and title
plt.title('Histogram of Age by Diagnosis Group', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Number of Patients', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()

# Show the plot
plt.show()
```

They have roughtly the same mean , median and standard deviation. The age distribution is similar across diagnosis groups.


```python
plt.figure(figsize=(10, 6))

box = plt.boxplot(
    [df[df['diagnosis'] == 0]['age'], df[df['diagnosis'] == 1]['age']],
    patch_artist=True,
    tick_labels=["No Alzheimer's (0)", "Alzheimer's (1)"]  # âœ… For Matplotlib 3.9+
)

# Custom colors
colors = ['#3498db', '#e74c3c']  # Blue, Red

# Set colors for box fill
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

# Customize median line
for median in box['medians']:
    median.set(color='black', linewidth=2)

# Add titles and labels
plt.title('Boxplot of Age by Diagnosis Group', fontsize=16)
plt.ylabel('Age', fontsize=14)
plt.xlabel('Diagnosis', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

df['gender_label'] = df['gender'].map({0: 'Male', 1: 'Female'})

gender_counts = df.groupby(['gender_label', 'diagnosis']).size().unstack()
gender_percent = gender_counts.div(gender_counts.sum(axis=1), axis=0) * 100

colors = ['#3498db', '#e74c3c']

gender_percent.plot(kind='bar', stacked=True, color=colors, figsize=(8, 6), edgecolor='black')

plt.title("Percentage of Alzheimer's Diagnosis by Gender", fontsize=16)
plt.xlabel("Gender", fontsize=14)
plt.ylabel("Percentage of Patients", fontsize=14)
plt.legend(["No Alzheimer's", "Alzheimer's"], loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=0)
plt.ylim(0, 100)

for i, (idx, row) in enumerate(gender_percent.iterrows()):
    plt.text(i - 0.15, row[0] / 2, f"{row[0]:.1f}%", color='white', fontweight='bold')
    plt.text(i - 0.15, row[0] + row[1] / 2, f"{row[1]:.1f}%", color='white', fontweight='bold')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Step 1: Map ethnicity codes to readable labels
df['ethnicity_label'] = df['ethnicity'].map({
    0: 'Caucasian',
    1: 'African American',
    2: 'Asian',
    3: 'Other'
})

# Step 2: Group by ethnicity and diagnosis, then get percentages
ethnicity_counts = df.groupby(['ethnicity_label', 'diagnosis']).size().unstack()
ethnicity_percent = ethnicity_counts.div(ethnicity_counts.sum(axis=1), axis=0) * 100

# Step 3: Plot
colors = ['#3498db', '#e74c3c']  # Blue = No Alzheimer's, Red = Alzheimer's
ethnicity_percent.plot(kind='bar', stacked=True, color=colors, figsize=(10, 6), edgecolor='black')

# Titles and formatting
plt.title("Percentage of Alzheimer's Diagnosis by Ethnicity", fontsize=16)
plt.xlabel("Ethnicity", fontsize=14)
plt.ylabel("Percentage of Patients", fontsize=14)
plt.legend(["No Alzheimer's", "Alzheimer's"], loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=15)
plt.ylim(0, 100)

# Step 4: Annotate bars with percentage labels
for i, (idx, row) in enumerate(ethnicity_percent.iterrows()):
    plt.text(i - 0.2, row[0] / 2, f"{row[0]:.1f}%", color='white', fontweight='bold')
    plt.text(i - 0.2, row[0] + row[1] / 2, f"{row[1]:.1f}%", color='white', fontweight='bold')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Risk factors and display labels
risk_factors = {
    'hypertension': 'Hypertension',
    'depression': 'Depression',
    'headinjury': 'Head Injury',
    'diabetes': 'Diabetes',
    'cardiovasculardisease': 'Cardiovascular Disease',
    'familyhistoryalzheimers': 'Family History',
    'smoking': 'Smoking'
}

colors = ['#3498db', '#e74c3c']  # Blue = No Alzheimer's, Red = Alzheimer's

# Create subplots: 2x4 grid but only fill 7
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.flatten()

# Loop and plot
for i, (col, label) in enumerate(risk_factors.items()):
    # Group and calculate percentages
    group = df.groupby([col, 'diagnosis']).size().unstack()
    percent = group.div(group.sum(axis=1), axis=0) * 100

    # Rename index
    if col == 'smoking':
        percent.index = ['Non-Smoker', 'Smoker']
    elif col == 'familyhistoryalzheimers':
        percent.index = ['No Family History', 'Has Family History']
    else:
        percent.index = ['No', 'Yes']

    # Plot
    percent.plot(kind='bar', stacked=True, color=colors, ax=axes[i], edgecolor='black')
    axes[i].set_title(f'Alzheimer\'s by {label}', fontsize=14)
    axes[i].set_ylabel('Percentage')
    axes[i].set_ylim(0, 100)
    axes[i].set_xticklabels(percent.index, rotation=0)
    axes[i].grid(axis='y', linestyle='--', alpha=0.6)
    axes[i].legend().set_visible(False)

    # Annotations
    for j, (idx, row) in enumerate(percent.iterrows()):
        axes[i].text(j - 0.15, row[0] / 2, f"{row[0]:.1f}%", color='white', fontweight='bold')
        axes[i].text(j - 0.15, row[0] + row[1] / 2, f"{row[1]:.1f}%", color='white', fontweight='bold')

# Remove the empty 8th plot
fig.delaxes(axes[-1])

# Add legend and title
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, ["No Alzheimer's", "Alzheimer's"], loc='lower center', ncol=2, fontsize=12)
plt.suptitle("Alzheimer's Diagnosis Breakdown by Risk Factors", fontsize=18, y=1.02)
plt.tight_layout(rect=[0, 0.05, 1, 0.95])
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Set style
sns.set(style="whitegrid")

# Variables to plot
lifestyle_vars = {
    'physicalactivity': 'Physical Activity (hrs/week)',
    'sleepquality': 'Sleep Quality (score)',
    'dietquality': 'Diet Quality (score)'
}

# Create figure and subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Color palette: Blue = No AD, Red = AD
palette = {0: '#3498db', 1: '#e74c3c'}

# Loop through variables and create boxplots
for ax, (col, label) in zip(axes, lifestyle_vars.items()):
    sns.boxplot(
        x='diagnosis',
        y=col,
        data=df,
        ax=ax,
        palette=palette
    )
    ax.set_title(f'{label} vs. Diagnosis', fontsize=14)
    ax.set_xlabel('Diagnosis')
    ax.set_ylabel(label)
    ax.set_xticklabels(['No Alzheimer\'s', 'Alzheimer\'s'])

# Overall title and layout
plt.suptitle("Healthy Lifestyle Factors vs. Alzheimer's Diagnosis", fontsize=18, y=1.05)
plt.tight_layout(rect=[0, 0.02, 1, 0.95])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Set seaborn style
sns.set(style="whitegrid")

# Define variables and readable labels
clinical_vars = {
    'cholesteroltotal': 'Total Cholesterol',
    'cholesterolldl': 'LDL Cholesterol',
    'cholesterolhdl': 'HDL Cholesterol',
    'cholesteroltriglycerides': 'Triglycerides',
    'systolicbp': 'Systolic BP',
    'diastolicbp': 'Diastolic BP'
}

# Create 2 rows x 3 columns subplot grid
fig, axes = plt.subplots(2, 3, figsize=(20, 10))
axes = axes.flatten()

# Define color palette
palette = {0: '#3498db', 1: '#e74c3c'}  # 0: No AD, 1: AD

# Loop through each variable
for i, (col, label) in enumerate(clinical_vars.items()):
    sns.boxplot(
        x='diagnosis',
        y=col,
        data=df,
        ax=axes[i],
        palette=palette
    )
    axes[i].set_title(f'{label} vs. Diagnosis', fontsize=14)
    axes[i].set_xlabel('Diagnosis')
    axes[i].set_ylabel(label)
    axes[i].set_xticklabels(['No Alzheimer\'s', 'Alzheimer\'s'])

# Final layout and title
plt.suptitle("Cholesterol and Blood Pressure Distributions by Alzheimer's Diagnosis", fontsize=18, y=1.03)
plt.tight_layout(rect=[0, 0.02, 1, 0.95])
plt.show()
```
